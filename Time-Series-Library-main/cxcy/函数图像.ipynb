```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc("font", family = 'SimHei')
#r0 = 0.1
r0 = 0.1851
u0 = 1 - r0
#p = 0.63
p = 0.667
# 定义函数
def f(x, r0, p):
    f = 1 / (1 + (1 / u0 - 1) * np.exp(p * x))
    #f = 1 - 1 / (1 + (1 / r0) * np.exp(-p * x))
    return f



# 生成x的值
x = np.linspace(0,15, 2000)  # 生成一个线性空间，从-10到10，共200个点

# 计算对应的y的值
y = f(x, r0, p)

# 绘制函数图像
plt.plot(x, y)

# 设置图表标题和坐标轴标签
plt.title('函数 y = x^2 的图像')
plt.xlabel('x')
plt.ylabel('y')

# 显示网格
plt.grid(True)

# 显示图像
plt.show()
```
