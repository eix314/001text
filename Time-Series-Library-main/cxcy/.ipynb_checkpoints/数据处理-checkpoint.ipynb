```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.preprocessing import MinMaxScaler

# 数据预处理
df['销售日期'] = pd.to_datetime(df['销售日期'])
df.set_index('销售日期', inplace=True)
data = df[['销量(千克)', '销售单价(元/千克)', '批发价格(元/千克)']].values
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# 定义时间步数
timesteps = 30

# 准备LSTM输入数据
X = []
y = []
for i in range(timesteps, data.shape[0]):
    X.append(scaled_data[i-timesteps:i, :])
    y.append(scaled_data[i, :])
X, y = np.array(X), np.array(y)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 构建LSTM模型
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(timesteps, scaled_data.shape[1]), dropout=0.2, recurrent_dropout=0.2))
model.add(LSTM(units=50, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(scaled_data.shape[1]))

# 编译模型
optimizer = Adam(learning_rate=0.003, decay=1e-6)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# 设置回调函数
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_loss')
]

# 训练模型
history = model.fit(X_train, y_train, epochs=150, batch_size=32, validation_data=(X_test, y_test), callbacks=callbacks)


# 评估模型
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Test RMSE: {rmse}')

# 可视化训练历史
plt.figure(figsize=(12, 4))

# 绘制训练损失
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='训练损失')
plt.plot(history.history['val_loss'], label='验证损失')
plt.legend()
plt.title('训练损失和验证损失')

# 显示图表
plt.show()
```

```python
def f_I_1(I_11, D_1, Q, mu_1):
    global D_2, S_1
    if mu_1 < 0.4:
        I_1 = 0
    elif Q > 0:
        I_1 = max(Q - D_1, 0)
        if I_1:
            D_2 = 0
            S_1 = D_1
        else:
            D_2 = min(D_1 - Q, D_2)
            S_1 = Q
    else :
        I_1 = max(I_11-D_1, 0)
        if I_1:
            D_2 = 0
            S_1 = D_1
        else:
            D_2 = min(D_1 - Q, D_2)
            S_1 = I_11
    return I_1

def f_I_2(I_22, I_11, D_2, Q, mu_2):
    global S_2
    if mu_2 < 0.2:
        I_2 = 0
    elif Q > 0:
        I_2 = max(I_11 + I_22 - D_2, 0)
        if I_2:
            S_2 = D_2
        else:
            S_2 = I_11 + I_22 
    else:
        I_2 = max(I_22 - D_2, 0)
        if I_2:
            S_2 = D_2
        else:
            S_2 = I_22
    return I_2


def f_mu_1(Q, I_11):
    global t_1
    if Q > 0:
        t_1=0
        mu_1 = 1 / (1 + (1 / mu_0 - 1) * np.exp(lambda_ * t_1))
    elif I_11 > 0 :
        t_1+=1
        mu_1 = 1 / (1 + (1 / mu_0 - 1) * np.exp(lambda_ * t_1))
    else:
        t_1 = np.inf
        mu_1 = 0
    return mu_1

def f_mu_2(Q, I_22):
    global t_1, t_2
    if Q > 0:
        t_2 = t_1 + 1
        mu_2 = 1 / (1 + (1 / mu_0 - 1) * np.exp(lambda_ * t_2))
    elif I_22 > 0:
        t_2 += 1
        mu_2 = 1 / (1 + (1 / mu_0 - 1) * np.exp(lambda_ * t_2))
    else:
        t_2 = np.inf
        mu_2 = 0
    return mu_2
```

```python
for i in range(len(D)):
    mu_11 = mu_1
    mu_22 = mu_2
    I_11 = I_1
    I_22 = I_2

    if mu_11 <= 0.4 or I_11+I_22 <= D[i]:
        Q = D[i]
    else:
        Q = 0

    mu_1 = f_mu_1(Q, I_11)
    mu_2 = f_mu_2(Q, I_22)

    if t_1 <= np.inf and t_2 <= np.inf:
        D_1 = d[i]
        #D_1 = D[i]*(1 - (1/(theta_**((m+1)/3)*np.exp((m-1)*(t_2-t_1)*lambda_/3)+1)))
        D_2 = d[i]*(1/(theta_**((4-m)/3)*np.exp((m-1)*(t_2-t_1)*lambda_/3)+1))
    elif t_1 == np.inf and t_2 == np.inf:
        D_1 = 0
        D_2 = 0
    elif t_2 == np.inf:
        D_1 = d
        D_2 = 0
    elif t_1 == np.inf:
        D_1 = 0
        D_2 = d
    
    # if t_1 <= np.inf and t_2 <= np.inf:
    #     D_1 = d[i]*(1 - (1/(theta_**((m+1)/3)*np.exp((m-1)*(t_2-t_1)*lambda_/3)+1)))
    #     D_2 = d[i]*(1/(theta_**((4-m)/3)*np.exp((m-1)*(t_2-t_1)*lambda_/3)+1))
    # elif t_1 == np.inf and t_2 == np.inf:
    #     D_1 = 0
    #     D_2 = 0
    # elif t_2 == np.inf:
    #     D_1 = d
    #     D_2 = 0
    # elif t_1 == np.inf:
    #     D_1 = 0
    #     D_2 = d


    I_1 = f_I_1(I_11, D_1, Q, mu_1)
    
    I_2 = f_I_2(I_22, I_11, D_2, Q, mu_2)

    if I_2:
        indicator_I_2 = 1
    else:
        indicator_I_2 = 0

    if Q:
        indicator_Q = 1
    else:
        indicator_Q = 0

    # S_1 = max(I_1,D_1)
    # S_2 = max(I_2,D_2)
    
    ls_S_1.append(S_1)
    ls_S_2.append(S_2)
    ls_Q.append(Q)
    ls_ir_I_2.append(indicator_I_2)
    ls_ir_Q.append(indicator_Q)
    ls_t_1.append(t_1)
    ls_t_2.append(t_2)
    ls_D_1.append(D_1)
    ls_D_2.append(D_2)
    ls_I_1.append(I_1)
    ls_I_2.append(I_2)

print(ls_S_1,
    ls_S_2,
    ls_D_1,
    ls_D_2,
    ls_I_1,
    ls_I_2,
    ls_Q,
    ls_ir_I_2,
    ls_ir_Q,
    ls_t_1,
    ls_t_2,sep = '\n\n')
```

```python
for t in range(len(D)):
    # 定义变量
    P = cp.Variable(2,pos=True)
    
    # 定义目标函数
    fun = (ls_S_1[t]*P[0] + ls_S_2[t]*P[1])*(1 - beta_ - ls_ir_I_2[t]*alpha_) - ls_ir_Q[t]*(C_F + ls_Q[t]*C_V[t]) - C_I
    objective = cp.Maximize(fun)
    
    # 定义约束条件
    if ls_t_1[t] <= np.inf and ls_t_2[t] <= np.inf:
        constraints = [P[1] <= P[0]*(theta_**((4-m)/3)*np.exp((1-m)*(t_2-t_1)*lambda_/3))]
    elif ls_t_1[t] == np.inf and ls_t_2[t] == np.inf:
        constraints = [P[1]==0,P[0]==0]
    elif ls_t_2[t] == np.inf:
        constraints = [P[1]==0]
    elif ls_t_1[t] == np.inf:
        constraints = [P[0]==0]
    
    # 构建优化问题
    prob = cp.Problem(objective, constraints)
    
    # 求解优化问题
    prob.solve(solver='GLPK_MI')
    
    # 输出结果
    ls_P.append(P.value)

print(ls_P)
```

```python
import numpy as np
from scipy.optimize import minimize

# 假设这些是已定义的参数和列表
# beta_, alpha_, C_F, C_V, C_I, theta_, lambda_, m, D, ls_S_1, ls_S_2, ls_ir_I_2, ls_ir_Q, ls_Q, ls_t_1, ls_t_2

ls_P = []

for t in range(len(D)):
    # 定义目标函数
    def objective(P):
        return -((ls_S_1[t]*P[0] + ls_S_2[t]*P[1]) * (1 - beta_ - ls_ir_I_2[t]*alpha_) 
                 - ls_ir_Q[t]*(C_F + ls_Q[t]*C_V[t]) - C_I)
    
    # 定义约束条件
    constraints = []
    if ls_t_1[t] <= np.inf and ls_t_2[t] <= np.inf:
        constraints.append({'type': 'ineq', 'fun': lambda P: P[1] - P[0]*(theta_**((4-m)/3)*np.exp((1-m)*(ls_t_2[t]-ls_t_1[t])*lambda_/3))})
    elif ls_t_1[t] == np.inf and ls_t_2[t] == np.inf:
        constraints.append({'type': 'eq', 'fun': lambda P: P[0]})
        constraints.append({'type': 'eq', 'fun': lambda P: P[1]})
    elif ls_t_2[t] == np.inf:
        constraints.append({'type': 'eq', 'fun': lambda P: P[1]})
    elif ls_t_1[t] == np.inf:
        constraints.append({'type': 'eq', 'fun': lambda P: P[0]})

    # # 定义约束条件
    # if ls_t_1[t] <= np.inf and ls_t_2[t] <= np.inf:
    #     constraints = [P[1] <= P[0]*(theta_**((4-m)/3)*np.exp((1-m)*(t_2-t_1)*lambda_/3))]
    # elif ls_t_1[t] == np.inf and ls_t_2[t] == np.inf:
    #     constraints = [P[1]==0,P[0]==0]
    # elif ls_t_2[t] == np.inf:
    #     constraints = [P[1]==0]
    # elif ls_t_1[t] == np.inf:
    #     constraints = [P[0]==0]
    
    # 定义变量边界
    bounds = [(0, C_V[t]*1.5), (0, C_V[t]*1.5)]  # P的每个元素都是非负的
    
    # 初始猜测
    x0 = [1, 1]
    
    # 求解优化问题
    result = minimize(objective, x0, method='SLSQP', bounds=bounds, constraints=constraints)
    
    # 输出结果
    if result.success:
        ls_P.append(result.x)
    else:
        ls_P.append(None)

print(ls_P)
```
